```python
result = [num for num in range(10)]
result
```

```python
## Squeare of number from o to 10 

lst = []
for n in range(11):
    val = n ** 2
    lst.append(val)
lst
```

```python
result = [num**2 for num in range(10)]
result
```

```python
## Squeare of number from 0 to 10  do this operation only for even numbers 

lst = []
for n in range(11):
    val = n ** 2
    if n % 2 == 0:
        lst.append(val)
lst
```
